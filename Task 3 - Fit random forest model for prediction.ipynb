## Task 3.1	

#### Fit a random forest model on “Train.csv” with a random state of 1, 500 estimators, a maximum depth of 3 and a maximum feature of 3. 

```python
rfcl = RandomForestClassifier(n_estimators = 500, max_depth=3, random_state=1, max_features=3)
rfcl = rfcl.fit(X_train, y_train)
```

```python
pdataT = pd.read_csv('C:\\Users\\stunn\\Notebook\\Capstone\\Test.csv')
pdataT.head()
```

```python
# Convert the relevant variables such as payment variables (Pay0-Pay6 and customer related variables) to categorical variables as appropriate.

payment_columns = ['SEX','EDUCATION','MARRIAGE','PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
pdataT[payment_columns] = pdataT[payment_columns].astype('category')
```

```python
pdataT.drop(['ID', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6'],axis=1,inplace=True)
pdataT.head()
```

```python
X_test = pdataT.iloc[:,:-1].values
y_test = pdataT['Default'].values
```

```python
y_pred = rfcl.predict(X_test) 
```

## Task 3.2	

#### Evaluate the confusion matrix, F1 scores and accuracy. Compare the random forest model with the logistic regression from Task 2. State your observations in a markdown cell in your Jupyter notebook.

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
print(metrics.classification_report(y_test, y_pred))
```
