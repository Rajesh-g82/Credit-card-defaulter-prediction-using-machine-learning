```python
# Convert the relevant variables such as payment variables (Pay0-Pay6 and customer related variables) to categorical variables as appropriate.

payment_columns = ['SEX','EDUCATION','MARRIAGE','PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
pdata[payment_columns] = pdata[payment_columns].astype('category')
```

```python
# Dropping irrelevant columns

pdata.drop(['ID', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6','PAY_AMT5'],axis=1,inplace=True)
pdata.head()
```

```python
X_train = pdata.iloc[:,0:-1].values
y_train = pdata['Default'].values
```

```python
X_train = preprocessing.normalize(X_train)
y_train = np.array(y_train)
```

```python
pdataT = pd.read_csv('C:\\Users\\stunn\\Notebook\\Capstone\\Test.csv')
pdataT.head()
```

```python
pdataT.drop(['ID', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6','PAY_AMT5'],axis=1,inplace=True)
pdataT.head()
```

```python
X_test = pdataT.iloc[:,:-1].values
y_test = pdataT['Default'].values
```

```python
X_test = preprocessing.normalize(X_test)
y_test = np.array(y_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes=2)
y_test = tensorflow.keras.utils.to_categorical(y_test,num_classes= 2)
y_train[0]
```

```python
model = Sequential()
```

```python
model.add(Input(shape=(16,)))
model.add(Dense(8, activation='relu', kernel_initializer='normal'))
model.add(Dense(8, activation='relu', kernel_initializer='normal'))
model.add(Dense(2, activation='softmax'))
```

```python
#model.compile(optimizer = sgd, loss='categorical_crossentropy', metrics=['accuracy'])
model.compile(optimizer = 'adam',loss='categorical_crossentropy')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 100, verbose = 1, batch_size=15)

#fit(self, x=None, y=None, batch_size=None, epochs=1, verbose=1, callbacks=None, validation_split=0.0, validation_data=None, shuffle=True, 
#class_weight=None, sample_weight=None, initial_epoch=0, steps_per_epoch=None, validation_steps=None)
```

```python
results = model.evaluate(X_test, y_test)
```

```python
print(model.metrics_names)
print(results) 
```

```python
Y_pred_cls = model.predict(X_test, verbose=0)
Y_pred_cls
```

```python
YYpred = np.argmax( Y_pred_cls, axis=1)
YYpred
```

```python
YYtest = np.argmax( y_test, axis=1) 
```

```python
YYtest
```

```python
print(metrics.classification_report(YYtest, YYpred))
```

```python
print(metrics.confusion_matrix(YYtest, YYpred))
```
